### **2. Könyvtárak importálása**




```
import io # Lehetővé teszi, hogy memóriában lévő adatfolyamokat (pl. letöltött fájlokat) fájlként kezeljünk Pythonban.
import requests # HTTP kéréseket küldhetünk, például adatfájlokat tölthetünk le egy weboldalról.
import pandas as pd # A Python egyik legfontosabb adatelemző könyvtára, amely táblázatos adatok (DataFrame) kezelésére szolgál.
import numpy as np # Numerikus számításokra és nagy méretű tömbök hatékony kezelésére használt alapvető tudományos Python könyvtár.
import matplotlib.pyplot as plt # A Python egyik legelterjedtebb grafikonrajzoló eszköze, amely lehetővé teszi különböző típusú vizualizációk létrehozását.

pd.set_option("display.max_columns", None) # Ez a beállítás biztosítja, hogy a pandas DataFrame-ek megjelenítésekor az összes oszlop látható legyen.
pd.set_option("display.float_format", lambda x: f"{x:.4f}") # Azt szabályozza, hogy a lebegőpontos számok négy tizedesjegyre kerekítve jelenjenek meg a DataFrame-ekben.
```

### **3. Paraméterek**

```
START_YEAR = 2013 # Meghatározzuk az elemzés kezdő évét, amelytől az iparági adatokat vizsgálni fogjuk.
END_YEAR_ARCHIVE = 2024 # Ez az utolsó év, amelynek adatai az archív Damodaran fájlokban találhatók.
CURRENT_YEAR = 2025 # Ez az aktuális év, amelyhez külön, friss adatfájl tartozik a Damodaran adatbázisban.

ARCHIVE_BASE = "https://pages.stern.nyu.edu/~adamodar/pc/archives/EVA{}.xls" # Ez egy URL sablon, amelynek segítségével programozottan letölthetjük az egyes évek archív EVA adatfájljait.
CURRENT_URL = "https://pages.stern.nyu.edu/~adamodar/pc/datasets/EVAGlobal.xls" # Ez a link az aktuális (legfrissebb) globális EVA adatfájlt tartalmazza, amelyet külön töltünk le.
```

### **4. Segédfüggvények a robusztus beolvasáshoz**

Az van, hogy Damodaran Professzor úr adatfájljai (excel) időben struktúrájukat tekintve változnak, évről évre nem egységes a forma. Ezért pontosan nem algorizmizálható a letöltés, hanem úgynevezett robusztus megoldással keresünk rá a szükséges adatokra (pl. megfelelő sheet, fejléc, ROC-WACC oszlopok).

```
# Excel sheet kiválasztása

def choose_industry_sheet(xls: pd.ExcelFile) -> str: # Ez a függvény kiválasztja azt az Excel munkalapot, amely az iparági átlagokat tartalmazza.
    if "Industry Averages" in xls.sheet_names: # Megvizsgáljuk, hogy létezik-e az Industry Averages nevű sheet.
        return "Industry Averages" # Ha létezik, akkor ezt használjuk, mert itt találhatók az iparági adatok.
    return xls.sheet_names[0] # Ha nem található ilyen nevű sheet, akkor az első munkalapot választjuk alapértelmezésként.

# Fejléc automatikus megtalálása

def find_header_row(xls: pd.ExcelFile, sheet: str) -> int: # Ez a függvény megkeresi, hogy az Excel fájlban melyik sor tartalmazza a tényleges oszlopfejléceket.
    raw = pd.read_excel(xls, sheet_name=sheet, header=None) # Az egész munkalapot beolvassuk fejléc nélkül, hogy minden sort vizsgálni tudjunk.
    hits = raw[raw.eq("Industry Name").any(axis=1)] # Az egész munkalapot beolvassuk fejléc nélkül, hogy minden sort vizsgálni tudjunk.
    if len(hits) == 0: # Ha nem találunk ilyen sort, akkor hibát jelzünk:
        raise ValueError(f"'Industry Name' header row not found in sheet={sheet}")
    return int(hits.index[0]) # Ha megtaláltuk, visszaadjuk annak a sornak az indexét, ahol a fejléc található.

# ROC−WACC oszlop automatikus azonosítása

def find_roc_wacc_column(df: pd.DataFrame) -> str: # Ez a függvény megkeresi azt az oszlopot, amely a ROC − WACC értékeket tartalmazza.
    candidates = [c for c in df.columns if ("ROC" in str(c)) and ("WACC" in str(c))] # Ez a függvény megkeresi azt az oszlopot, amely a ROC − WACC értékeket tartalmazza.
    if not candidates: # Ha nem találunk ilyen oszlopot, hibát jelez a kód:
        raise ValueError(f"ROC-WACC column not found. Columns: {list(df.columns)}")
    return candidates[0] # Ha több találat van, akkor az első megfelelő oszlopot választjuk.

# Ez a függvény felelős az egyes évek Damodaran EVA adatainak letöltéséért, tisztításáért és egységes formátumba alakításáért, hogy azokat később időbeli elemzésre fel tudjuk használni.

def load_eva_excel_from_url(url: str, year: int) -> pd.DataFrame: # Ez a függvény egy megadott URL-ről letölt egy Damodaran EVA Excel fájlt, majd az adatokat egy pandas DataFrame formájában adja vissza.
    r = requests.get(url, timeout=60) # Ez a függvény egy megadott URL-ről letölt egy Damodaran EVA Excel fájlt, majd az adatokat egy pandas DataFrame formájában adja vissza.
    r.raise_for_status() # HTTP kéréssel letöltjük az Excel fájlt a megadott URL-ről.

    xls = pd.ExcelFile(io.BytesIO(r.content)) # A letöltött fájl bináris tartalmát memóriában Excel fájlként megnyitjuk.
    sheet = choose_industry_sheet(xls) # Kiválasztjuk azt a munkalapot, amely az iparági adatokat tartalmazza.
    header_row = find_header_row(xls, sheet) # Automatikusan meghatározzuk, melyik sor tartalmazza az oszlopneveket.

    df = pd.read_excel(xls, sheet_name=sheet, skiprows=header_row) # Beolvassuk az Excel adatokat a pandas DataFrame-be, kihagyva a fejléc előtti sorokat.
    roc_col = find_roc_wacc_column(df) # Megkeressük azt az oszlopot, amely a ROC−WACC mutatót tartalmazza.

    out = df[["Industry Name", roc_col]].dropna().copy() # Megkeressük azt az oszlopot, amely a ROC−WACC mutatót tartalmazza.
    out.columns = ["name", "roc_wacc"] # Egységes oszlopnevekre nevezzük át a kiválasztott adatokat.
    out["year"] = year # Hozzáadunk egy új oszlopot, amely jelzi, hogy az adatok melyik évből származnak.

    out["roc_wacc"] = pd.to_numeric(out["roc_wacc"], errors="coerce") # Hozzáadunk egy új oszlopot, amely jelzi, hogy az adatok melyik évből származnak.
    out = out.dropna(subset=["roc_wacc"]) # A ROC−WACC értékeket numerikus formátumba alakítjuk, a hibás értékeket hiányzóként kezelve.

    return out[["year", "name", "roc_wacc"]] # A függvény végül visszaadja az év, iparág és ROC−WACC értékeket tartalmazó tisztított adatokat.
```

### **5. Archív adatok betöltése 2013 - 2024**

Ebben a cellában ciklussal végigmegyünk az archív EVA fájlokon (2013–2024), minden évből betöltjük az iparági ROC−WACC adatokat, majd ezeket egy közös táblába egyesítjük.

```
frames = [] # Létrehozunk egy üres listát, amelyben az egyes évek adatkereteit fogjuk eltárolni.

for y in range(START_YEAR, END_YEAR_ARCHIVE + 1): # Egy ciklussal végigmegyünk az összes archív éven a kezdőévtől az utolsó archív évig.
    yy = str(y)[-2:] # Az aktuális év utolsó két számjegyét kinyerjük, mert az archív fájlnevek ezt a formátumot használják.
    url = ARCHIVE_BASE.format(yy) # Az archív EVA fájl URL-jét összeállítjuk az évszám alapján.
    df_year = load_eva_excel_from_url(url, y) # Letöltjük és feldolgozzuk az adott év EVA Excel fájlját a korábban definiált függvény segítségével.
    frames.append(df_year) # Az adott évből származó DataFrame-et hozzáadjuk a listához.

industry_all = pd.concat(frames, ignore_index=True) # Az összes éves adatkeretet egyetlen nagy táblává fűzzük össze.

industry_all.groupby("year").size() # Az összes éves adatkeretet egyetlen nagy táblává fűzzük össze.
```

### **6. A 2025. évi adatok hozzáadása**

Ebben a lépésben a legfrissebb (2025-ös) EVA adatokat töltjük be a Damodaran adatbázisból (mivel azokat máshol tárolta professzor úr, mint az azt megelőzőeket), majd hozzáadjuk az archív évekből már összeállított iparági adatkészlethez.

```
df_2025 = load_eva_excel_from_url(CURRENT_URL, CURRENT_YEAR) # A legfrissebb EVA adatfájlt letöltjük a Damodaran adatbázisból, majd a korábban definiált függvénnyel DataFrame formátumba alakítjuk.

industry_all = pd.concat([industry_all, df_2025], ignore_index=True) # A 2025-ös adatokat hozzáfűzzük a korábbi évek iparági adatokat tartalmazó táblájához.

industry_all.groupby("year").size() # A 2025-ös adatokat hozzáfűzzük a korábbi évek iparági adatokat tartalmazó táblájához.
```

### **7. Évenkénti Top10 / Worst 10 befutók kiválasztása ROC-WACC alapján**

Ebben a lépésben minden évre kiszámítjuk az iparágak ROC−WACC rangsorát, majd kiválasztjuk az adott év Top 10 és Worst 10 iparágait.

```
industry_all["rank_top"] = industry_all.groupby("year")["roc_wacc"].rank(method="first", ascending=False) # Az iparágakat minden évben ROC−WACC érték alapján rangsoroljuk úgy, hogy a legmagasabb érték kapja az első helyet.
industry_all["rank_bottom"] = industry_all.groupby("year")["roc_wacc"].rank(method="first", ascending=True) # Ugyanezt a rangsort kiszámítjuk fordított sorrendben is, hogy az alacsonyabb ROC−WACC értékek kerüljenek előre.

top10_df = industry_all[industry_all["rank_top"] <= 10].copy() # Kiválasztjuk azokat az iparágakat, amelyek az adott évben a legjobb tíz ROC−WACC érték közé kerültek.
worst10_df = industry_all[industry_all["rank_bottom"] <= 10].copy() # Kiválasztjuk azokat az iparágakat, amelyek az adott évben a tíz legalacsonyabb ROC−WACC érték közé kerültek.
```

### **8. Top10 - teljes idősor**

Ebben a lépésben kiválasztjuk azokat az iparágakat, amelyek valaha bekerültek az éves Top 10 mezőnybe, majd ezek teljes idősorát előkészítjük a vizualizációhoz.

```
top_names = sorted(top10_df["name"].unique()) # Összegyűjtjük és ábécé sorrendbe rendezzük azoknak az iparágaknak a nevét, amelyek bármely évben bekerültek a Top 10-be.

top_plot_df = industry_all[industry_all["name"].isin(top_names)].copy() # Az eredeti adatbázisból kiválasztjuk azoknak az iparágaknak az összes adatát, amelyek valaha Top 10-be kerültek.
top_plot_df = top_plot_df.sort_values(["name", "year"]).reset_index(drop=True) # Az adatokat iparág és év szerint rendezzük, majd újraindexeljük a táblát.

top_label_map = {} # Létrehozunk egy üres szótárat, amelyben az iparágakhoz tartozó sorszámokat és legend címkéket fogjuk tárolni.
for i, name in enumerate(top_names, start=1):  # Végigmegyünk az összes Top 10 iparágon, és mindegyikhez egy sorszámot rendelünk.
    top_label_map[name] = f"{i}. {name}" # Végigmegyünk az összes Top 10 iparágon, és mindegyikhez egy sorszámot rendelünk.

top_plot_df["label_id"] = top_plot_df["name"].map(lambda x: top_label_map[x].split(".")[0]) # Az iparágakhoz tartozó sorszámot külön oszlopban eltároljuk, hogy a grafikonon azonosításra használhassuk.

top_plot_df.head() # Az iparágakhoz tartozó sorszámot külön oszlopban eltároljuk, hogy a grafikonon azonosításra használhassuk.
```

### **9. Top10 grafikon**

Ebben a lépésben elkészítjük a publikációra szánt idősoros grafikonunkat, amely bemutatja azoknak az iparágaknak a ROC−WACC alakulását, amelyek valaha bekerültek az éves Top 10 mezőnybe.


```
plt.figure(figsize=(16, 10)) # Létrehozunk egy új ábrát, amelyen a grafikon megjelenik, és beállítjuk a publikációhoz megfelelő méretet.
colors = plt.cm.tab20.colors # Kiválasztjuk a Matplotlib egyik előre definiált színpalettáját, hogy az egyes iparágakat különböző színekkel jelenítsük meg.

for i, name in enumerate(top_names): # Végigmegyünk az összes iparágon, amely valaha bekerült a Top 10-be.
    df_i = top_plot_df[top_plot_df["name"] == name].sort_values("year") # Kiválasztjuk az adott iparág teljes idősorát, majd időrendbe rendezzük.
    color = colors[i % len(colors)] # Kiválasztjuk az adott iparág teljes idősorát, majd időrendbe rendezzük.

    plt.plot( # Elkezdjük kirajzolni az adott iparág idősoros vonaldiagramját.
        df_i["year"], # Az x tengelyre az éveket helyezzük.
        df_i["roc_wacc"] * 100, # Az y tengelyre a ROC−WACC értéket tesszük, százalékos formára átszámítva.
        marker="o", # A vonaldiagram pontjait kör alakú jelölőkkel jelenítjük meg.
        linewidth=2.2, # A vonal vastagságát beállítjuk, hogy jobban látható legyen.
        color=color, # Az adott iparághoz tartozó színt alkalmazzuk a vonalra.
        label=top_label_map[name] # A grafikon jelmagyarázatában (legend) az iparág sorszámmal ellátott nevét jelenítjük meg.
    )


    last_row = df_i.iloc[-1] # Kiválasztjuk az adott iparág idősorának utolsó (legfrissebb) megfigyelését.
    plt.text( # Szöveges annotációt helyezünk el a grafikonon.
        last_row["year"] + 0.05, # A szöveget kicsit jobbra helyezzük az utolsó adatponttól.
        last_row["roc_wacc"] * 100, # A szöveg függőleges pozícióját az utolsó ROC−WACC értékhez igazítjuk.
        top_label_map[name].split(".")[0], # Csak az iparág sorszámát írjuk ki az utolsó pont mellé.
        fontsize=9, # Beállítjuk a felirat betűméretét.
        color=color, # A felirat színe megegyezik az iparág vonalának színével.
        va="center" # A szöveget függőlegesen középre igazítjuk az adatpont mellett.
    )

plt.axhline(y=0, color="black", linestyle="--", linewidth=1) # Kirajzolunk egy vízszintes referencia-vonalat a 0 értéknél, amely segít vizuálisan elkülöníteni a pozitív és negatív ROC−WACC tartományt.

plt.title("Iparágak, amelyek valaha Top 10-be kerültek – ROC−WACC alakulás (%)", fontsize=16) # Beállítjuk a grafikon címét, amely röviden összefoglalja a vizualizáció tartalmát.
plt.xlabel("Év", fontsize=12) # Megadjuk az x tengely feliratát, amely az idődimenziót jelöli.
plt.ylabel("ROC−WACC (%)", fontsize=12) # Beállítjuk az y tengely feliratát, amely a százalékos ROC−WACC értéket mutatja.

plt.legend( # Létrehozzuk a grafikon jelmagyarázatát
    title="Top mezőny", # A jelmagyarázathoz egy címet adunk.
    bbox_to_anchor=(1.02, 1), # A jelmagyarázatot kissé a grafikon jobb oldalán kívülre helyezzük.
    loc="upper left", # A jelmagyarázat horgonyzási pontját a bal felső sarokhoz igazítjuk.
    borderaxespad=0, # Eltávolítjuk a grafikon és a jelmagyarázat közötti extra térközt.
    fontsize=9 # Beállítjuk a jelmagyarázat betűméretét.
)

plt.grid(True, linestyle=":", alpha=0.6) # Finom, pontozott rácsvonalakat adunk a grafikonhoz, hogy könnyebb legyen az értékek vizuális követése.
plt.tight_layout() # Automatikusan optimalizáljuk az elrendezést, hogy a tengelyek, címek és a jelmagyarázat ne lógjanak ki a képből.
plt.show() # Tátááám! Megjelenítjük a kész grafikont a notebookban.
```

### **10. Worst10 - teljes idősor minden olyan iparágra, ami a vizsgált időszakban bármikor a legrosszabb 10 közé került ROC - WACC alapon**

Mivel ennek a logikája megegyezik a Top10-nél olvasottakkal (8. cella), ezért ide már nem írok ki magyarázatot.


```
worst_names = sorted(worst10_df["name"].unique())

worst_plot_df = industry_all[industry_all["name"].isin(worst_names)].copy()
worst_plot_df = worst_plot_df.sort_values(["name", "year"]).reset_index(drop=True)

worst_label_map = {}
for i, name in enumerate(worst_names, start=1):
    worst_label_map[name] = f"{i}. {name}"

worst_plot_df["label_id"] = worst_plot_df["name"].map(lambda x: worst_label_map[x].split(".")[0])

worst_plot_df.head()
```

### **11. Worst10 - megjelenítés**

Mivel ennek a logikája megegyezik a Top10-nél olvasottakkal (9. cella), ezért ide már nem írok ki magyarázatot.


```
plt.figure(figsize=(16, 10))

colors = plt.cm.tab20.colors

for i, name in enumerate(worst_names):
    df_i = worst_plot_df[worst_plot_df["name"] == name].sort_values("year")
    color = colors[i % len(colors)]

    plt.plot(
        df_i["year"],
        df_i["roc_wacc"] * 100,
        marker="o",
        linewidth=2.2,
        color=color,
        label=worst_label_map[name]
    )

    # utolsó pont mellé kiírjuk a sorszámot
    last_row = df_i.iloc[-1]
    plt.text(
        last_row["year"] + 0.05,
        last_row["roc_wacc"] * 100,
        worst_label_map[name].split(".")[0],
        fontsize=9,
        color=color,
        va="center"
    )

plt.axhline(y=0, color="black", linestyle="--", linewidth=1)

plt.title("Iparágak, amelyek valaha Worst 10-be kerültek – ROC−WACC alakulás (%)", fontsize=16)
plt.xlabel("Év", fontsize=12)
plt.ylabel("ROC−WACC (%)", fontsize=12)

plt.legend(
    title="Worst mezőny",
    bbox_to_anchor=(1.02, 1),
    loc="upper left",
    borderaxespad=0,
    fontsize=9
)

plt.grid(True, linestyle=":", alpha=0.6)
plt.tight_layout()
plt.show()
```

### **12. Évenkénti Top10 / Worst10 tábla előkészítése**

Ebben a lépésben minden évre kiszámoljuk az iparágak ROC−WACC rangsorát, majd megjelöljük, hogy az adott iparág az adott évben a Top 10 vagy a Worst 10 csoportba tartozik-e, és létrehozunk egy százalékos ROC−WACC oszlopot a későbbi elemzéshez.

```
industry_all["rank_top"] = industry_all.groupby("year")["roc_wacc"].rank(method="first", ascending=False) # Minden évben kiszámoljuk az iparágak rangsorát ROC−WACC alapján úgy, hogy a legmagasabb érték kapja az első helyet.
industry_all["rank_bottom"] = industry_all.groupby("year")["roc_wacc"].rank(method="first", ascending=True) # Ugyanígy kiszámoljuk az iparágak rangsorát, de itt a legalacsonyabb ROC−WACC érték kapja az első helyet.

industry_all["group_flag"] = "" # Létrehozunk egy új oszlopot, amelyben később jelölni fogjuk, hogy egy iparág az adott évben Top 10 vagy Worst 10 csoportba tartozik-e.

industry_all.loc[industry_all["rank_top"] <= 10, "group_flag"] = "Top10" # Azokat a sorokat, ahol az iparág a legjobb tíz között van, a Top10 címkével látjuk el.
industry_all.loc[industry_all["rank_bottom"] <= 10, "group_flag"] = "Worst10" # Azokat a sorokat, ahol az iparág a legrosszabb tíz között van, a Worst10 címkével jelöljük.


industry_all["roc_wacc_pct"] = industry_all["roc_wacc"] * 100 # A ROC−WACC értéket százalékos formára alakítjuk, hogy a későbbi vizualizációkban és BI eszközökben könnyebben kezelhető legyen.

industry_all.head() # Megjelenítjük az adatkeret első néhány sorát ellenőrzés céljából.
```

### **13. Évenkénti Top10 és Worst10 külön táblák**

Ebben a lépésben külön táblákba rendezzük az adott évek Top10 és Worst10 iparágait, ami megkönnyíti a későbbi vizualizációt és a BI eszközökben történő elemzést. Ezek a táblák különösen hasznosak interaktív dashboardokhoz (pl. Looker Studio), ahol az elemző év szerint szűrve gyorsan áttekintheti az adott időszak legjobb és legrosszabb iparágait.

```
top10_by_year = industry_all[industry_all["rank_top"] <= 10].copy() # Kiválasztjuk azokat a sorokat, ahol az iparág az adott évben a Top 10-be került.
worst10_by_year = industry_all[industry_all["rank_bottom"] <= 10].copy() # Kiválasztjuk azokat a sorokat, ahol az iparág az adott évben a 10 legrosszabb között szerepelt.

top10_by_year = top10_by_year.sort_values(["year", "rank_top"]) # A Top 10 táblát év és rangsor szerint rendezzük.
worst10_by_year = worst10_by_year.sort_values(["year", "rank_bottom"]) # A Worst 10 táblát év és rangsor szerint rendezzük.

top10_by_year.head(20) # Megjelenítjük a Top 10 tábla első néhány sorát ellenőrzés céljából.
```

### **14. Teljes idősor azoknak, akik valaha (a vizsgált időintervallumon belül) a Top10 / Worst10 mezőnyben voltak**

Ebben a lépésben nemcsak azokat az éveket tartjuk meg, amikor egy iparág Top 10 vagy Worst 10 volt, hanem az adott iparág teljes történetét is, így később teljes idősoros elemzést tudunk készíteni róluk.

Ez a megközelítés segít megkülönböztetni két kérdést:

1.  Kik voltak az adott évben a legjobbak és legrosszabbak?
2.  Hogyan alakult ezeknek az iparágaknak a teljes pályája az időben?


```
top_names = sorted(top10_by_year["name"].unique()) # Összegyűjtjük és ábécé sorrendbe rendezzük azoknak az iparágaknak a nevét, amelyek bármely évben bekerültek a Top 10-be.
worst_names = sorted(worst10_by_year["name"].unique()) # Összegyűjtjük és ábécé sorrendbe rendezzük azoknak az iparágaknak a nevét, amelyek bármely évben bekerültek a Worst 10-be.

top10_full_history = industry_all[industry_all["name"].isin(top_names)].copy() # Az eredeti adatbázisból kiválasztjuk azoknak az iparágaknak az összes megfigyelését, amelyek valaha Top 10-esek voltak.
worst10_full_history = industry_all[industry_all["name"].isin(worst_names)].copy() # Az eredeti adatbázisból kiválasztjuk azoknak az iparágaknak az összes megfigyelését, amelyek valaha Worst 10-esek voltak.

top10_full_history = top10_full_history.sort_values(["name", "year"]).reset_index(drop=True) # A Top 10-es iparágak teljes idősorát iparág és év szerint rendezzük, majd újraindexeljük.
worst10_full_history = worst10_full_history.sort_values(["name", "year"]).reset_index(drop=True) # A Worst 10-es iparágak teljes idősorát szintén iparág és év szerint rendezzük, majd újraindexeljük.

top10_full_history.head() # Ellenőrzésképpen megjelenítjük a Top 10-es iparágak teljes idősorának első néhány sorát.
```
